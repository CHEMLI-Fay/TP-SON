## 2.1/ Coding the beamformer filters and analyzing their properties

**These first questions have to be prepared before the practical session**

### 2) Write the position $z_n$ as a function of $n$ and interspace $d$. As a convention, the first microphone number is selected as $0$.

One can write :
$$z_n = (n - (N+1)/2) . d$$

### 3) Propose a function `beam_filter` returning the filter frequency response for one microphone number `mic_nb`. 

```python
def beam_filter(array, freq_vector, theta0=0, mic_nb: int = 0):
    """Compute the filter frequency response of a DSB beamformer for one microphone
    Args:
        array (array_server obj): array structure controlling the acquisition system.
        freq_vector (np.array): frequency vector. 
        theta0 (int, optional): focusing angular direction (in degrees). Defaults to 0.
        mic_nb (int, optional): microphone number. Defaults to 0.
    Returns:
        np.array: the filter frequency response. Shape is (len(freq_vector),).
    """

    c = 343  # Speed of sound in m/s
    
    # Microphone position x (origin at center of array)
    x = (mic_nb - (N - 1) / 2) * d
    
    # Convert angle from degrees to radians
    theta0_rad = np.deg2rad(theta0)
    
    # Filter's frequency response: exp(-j * k * x * cos(θ₀))
    # where k = 2πf/c
    return np.exp(-1j * 2 * np.pi * freq_vector * x * np.cos(theta0_rad) / c)
```

### 4) Plot the two frequency responses obtained for two filters associated to two different microphone outputs when $\theta_0=0^\circ$ and for frequencies between $0$ and $5$kHz. Explain the effect of these filters on the signals.

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters we need
c = 343  # Speed of sound in m/s

# Frequency vector from 0 to 5 kHz
freq_vector = np.linspace(0, 5000, 1000)

# Theta0 = 0 degrees
theta0 = 0

# Choose TWO different microphones to compare
mic_nb_1 = 0  # First microphone
mic_nb_2 = 3 # Last microphone

# Compute the two filter responses
W1 = beam_filter(antenne, freq_vector, theta0=theta0, mic_nb=mic_nb_1)
W2 = beam_filter(antenne, freq_vector, theta0=theta0, mic_nb=mic_nb_2)

# Calculate positions for information
x1 = (mic_nb_1 - (N - 1) / 2) * d
x2 = (mic_nb_2 - (N - 1) / 2) * d


fig, axes = plt.subplots(2, 1, figsize=(12, 8))

# Plot 1: Magnitude Response
axes[0].plot(freq_vector, np.abs(W1), label=f'Mic {mic_nb_1} (x={x1*100:.1f} cm)', 
             linewidth=2.5, color='blue')
axes[0].plot(freq_vector, np.abs(W2), label=f'Mic {mic_nb_2} (x={x2*100:.1f} cm)', 
             linewidth=2.5, color='red')
axes[0].set_xlabel('Frequency (Hz)', fontsize=12)
axes[0].set_ylabel('Magnitude |W(f)|', fontsize=12)
axes[0].set_title('Magnitude Response of Beamformer Filters (θ₀ = 0°)', 
                  fontsize=13, fontweight='bold')
axes[0].grid(True, alpha=0.3)
axes[0].legend(fontsize=11)
axes[0].set_xlim([0, 5000])
axes[0].set_ylim([0.99, 1.01])

# Plot 2: Phase Response
axes[1].plot(freq_vector, np.unwrap(np.angle(W1)), label=f'Mic {mic_nb_1} (x={x1*100:.1f} cm)', 
             linewidth=2.5, color='blue')
axes[1].plot(freq_vector, np.unwrap(np.angle(W2)), label=f'Mic {mic_nb_2} (x={x2*100:.1f} cm)', 
             linewidth=2.5, color='red')
axes[1].set_xlabel('Frequency (Hz)', fontsize=12)
axes[1].set_ylabel('Phase ∠W(f) (radians)', fontsize=12)
axes[1].set_title('Phase Response of Beamformer Filters (θ₀ = 0°)', 
                  fontsize=13, fontweight='bold')
axes[1].grid(True, alpha=0.3)
axes[1].legend(fontsize=11)
axes[1].set_xlim([0, 5000])

plt.tight_layout()
plt.show()

# Calculate and display the time delays
theta0_rad = np.deg2rad(theta0)
tau1 = -x1 * np.cos(theta0_rad) / c
tau2 = -x2 * np.cos(theta0_rad) / c

```

## 2.2/ Using the filters : coding of the beamforming
Basically, the beamforming algorithm is the following :

- (a) acquire an audio frame
- (b) compute the corresponding FFT
- (c) analyze the FFT to define which frequency(ies) you would like to localize
- (d) restrict the FFT to the frequencies of interest
- (e) for one given ✓ 0 , for the frequencies selected before, and for each microphone :
  — compute the corresponding filters frequency responses with the beam_filter func-
tion
  — apply these filters to the microphone outputs
- (f) compute the beamformer output associated to the angular polarization ✓ 0
- (g) repeat all these last steps for each ✓ 0 you want to test
- (h) finally, decide of the angular position of the source by detecting for which ✓ 0 the beam-
former output is maximum.

### 6) Step (a) and (b) : After acquiring an audio buffer, compute its FFT in an array $M_{fft}$. Plot the result of this analysis as a function of the frequency when emitting a pure sine tone with a frequency $F_0 = 1$kHz.

```python
import numpy as np
import matplotlib.pyplot as plt

# Step (a): Acquire an audio buffer
print("Acquiring audio buffer...")
m = antenne.read()
print(f"Buffer shape: {m.shape}")  # Should be (BLK, N) = (2048, 8)

# Step (b): Compute FFT for each microphone
M_fft = np.fft.fft(m, axis=0)  # FFT along time axis (axis 0)
print(f"M_fft shape: {M_fft.shape}")  # Should be (2048, 8)

# Create frequency vector
freq_vector = np.fft.fftfreq(BLK, 1/Fs)

# Only keep positive frequencies (first half)
n_positive = BLK // 2
freq_positive = freq_vector[:n_positive]
M_fft_positive = M_fft[:n_positive, :]

print(f"Frequency range: {freq_positive[0]:.1f} Hz to {freq_positive[-1]:.1f} Hz")
print(f"Frequency resolution: {freq_positive[1] - freq_positive[0]:.2f} Hz")

# Plot the FFT magnitude for all microphones
fig, axes = plt.subplots(2, 1, figsize=(14, 10))

# Define colors for each microphone
colors = ['red', 'blue', 'green', 'orange', 'purple', 'brown', 'pink', 'gray']

# Plot 1: Full frequency range (0 to 10 kHz)
for i in range(N):
    axes[0].plot(freq_positive, np.abs(M_fft_positive[:, i]), 
                 label=f'Mic {i}', color=colors[i], alpha=0.7, linewidth=1.5)

axes[0].set_xlabel('Frequency (Hz)', fontsize=12)
axes[0].set_ylabel('Magnitude', fontsize=12)
axes[0].set_title('FFT Magnitude - All Microphones (Full Range)', fontsize=13, fontweight='bold')
axes[0].set_xlim([0, 10000])
axes[0].grid(True, alpha=0.3)
axes[0].legend(loc='upper right', fontsize=9)

# Plot 2: Zoomed around F0 = 1 kHz
F0 = 1000  # Expected frequency
zoom_range = 200  # Hz on each side

for i in range(N):
    axes[1].plot(freq_positive, np.abs(M_fft_positive[:, i]), 
                 label=f'Mic {i}', color=colors[i], alpha=0.7, linewidth=2)

axes[1].set_xlabel('Frequency (Hz)', fontsize=12)
axes[1].set_ylabel('Magnitude', fontsize=12)
axes[1].set_title(f'FFT Magnitude - Zoomed around F₀ = {F0} Hz', fontsize=13, fontweight='bold')
axes[1].set_xlim([F0 - zoom_range, F0 + zoom_range])
axes[1].grid(True, alpha=0.3)
axes[1].legend(loc='upper right', fontsize=9)
axes[1].axvline(x=F0, color='black', linestyle='--', linewidth=1, alpha=0.5, label='Expected F₀')

plt.tight_layout()
plt.show()

# Find the peak frequency (should be around 1 kHz)
# Average across all microphones to find the peak
M_fft_avg = np.mean(np.abs(M_fft_positive), axis=1)
peak_idx = np.argmax(M_fft_avg)
peak_freq = freq_positive[peak_idx]


# Display FFT values at the peak for all microphones
print(f"\nFFT values at peak frequency ({peak_freq:.2f} Hz):")
for i in range(N):
    mag = np.abs(M_fft_positive[peak_idx, i])
    phase = np.angle(M_fft_positive[peak_idx, i])
    print(f"  Mic {i}: Magnitude = {mag:.1f}, Phase = {phase:.3f} rad ({np.degrees(phase):.1f}°)")

print("="*70)
```

### 7) Step (c) and (d) : Among all the frequencies you obtained from the FFT, select the one corresponding to the source frequency. Give its exact value and index $k_0$ in the frequency array, and collect the corresponding FFT values of each microphone outputs in one vector $\texttt{M}$ of length $N$.

```python
import numpy as np

# Step (c): Find the frequency corresponding to the source (F0 = 1 kHz)

# Method 1: Find the peak in the averaged FFT across all microphones
M_fft_avg = np.mean(np.abs(M_fft_positive), axis=1)
k0 = np.argmax(M_fft_avg)
f0_detected = freq_positive[k0]


# Step (d): Collect the FFT values at frequency k0 for all microphones
# M is a vector of length N containing the complex FFT values
M = M_fft_positive[k0, :]

# Display the collected FFT values
print(f"\n" + "-"*70)
print(f"FFT values at k₀ = {k0} (f = {f0_detected:.2f} Hz):")
print("-"*70)
print(f"{'Mic':<5} {'Real':<15} {'Imag':<15} {'Magnitude':<15} {'Phase (rad)':<15} {'Phase (°)':<10}")
print("-"*70)

for i in range(N):
    real_part = np.real(M[i])
    imag_part = np.imag(M[i])
    magnitude = np.abs(M[i])
    phase_rad = np.angle(M[i])
    phase_deg = np.degrees(phase_rad)
    print(f"{i:<5} {real_part:<15.2f} {imag_part:<15.2f} {magnitude:<15.2f} {phase_rad:<15.3f} {phase_deg:<10.1f}")
```

### 8) Step (e) : In a loop among all microphones, compute each filters for the position $\theta_0$ and for the frequency value you obtained in the previous step. Apply then these filters to the array $\texttt{M}$ defined before.

```python
import numpy as np

# Choose a steering angle theta0 to test
theta0 = 45  # You can change this to any angle you want to test

# Create an array to store the filtered outputs
M_filtered = np.zeros(N, dtype=complex)


for n in range(N):
    # Compute the filter for microphone n at frequency f0_detected
    # We only need the filter at ONE frequency (f0_detected)
    W_n = beam_filter(antenne, np.array([f0_detected]), theta0=theta0, mic_nb=n)
    
    # W_n is an array of length 1, so extract the single value
    W_n_value = W_n[0]
    
    # Apply the filter to the FFT value M[n]
    M_filtered[n] = W_n_value * M[n]
    
    # Display the results
    print(f"Mic {n}: W[{n}] = {np.abs(W_n_value):.4f} ∠{np.degrees(np.angle(W_n_value)):7.2f}°  |  "
          f"M[{n}] = {np.abs(M[n]):8.1f} ∠{np.degrees(np.angle(M[n])):7.1f}°  →  "
          f"M_filt[{n}] = {np.abs(M_filtered[n]):8.1f} ∠{np.degrees(np.angle(M_filtered[n])):7.1f}°")

print("-"*70)
```

### 9) Step (f): Combine then the filters outputs to form the beamformer output $Y_{\theta_0}[k_0]$. *$Y_{\theta_0}[k_0]$ is obviously a complex value which corresponds to the frequency contribution of the source to the $k_0^{\text{th}}$ frequency component of the beamformer output when focalized in the direction $\theta_0$.* Compute then the corresponding power $P(\theta_0)$ at $k_0$ of the beamformer output.

```python
import numpy as np

print("="*70)
print("STEP (f): Combine Filtered Outputs and Compute Power")
print("="*70)

# Combine all filtered outputs by summing them
Y_theta0 = np.sum(M_filtered)

# Compute the power P(θ₀) at k₀
P_theta0 = np.abs(Y_theta0)**2

print("Power of beamformer output:")
print(f"P(θ₀) = |Y_θ₀[k₀]|² = {P_theta0:.2f}")





#P(θ₀) = |Y_θ₀[k₀]|²
```

### 10) For a direction $\theta_0$ of your choice, compute $P(\theta_0)$ for (i) a source emitting from a direction close to $\theta_0$, or (ii) far from it. Compare the two values.

```python
import numpy as np

print("="*70)
print("STEP 10: Compare Power for Different Source Positions")
print("="*70)

# Define two steering angles to test
theta0_close = -135  # Close to actual source direction
theta0_far = 0       # Far from actual source direction (or pick 90, 45, etc.)


# Compute filters and apply them for theta0_close
M_filtered_close = np.zeros(N, dtype=complex)
for n in range(N):
    W_n = beam_filter(antenne, np.array([f0_detected]), theta0=theta0_close, mic_nb=n)
    M_filtered_close[n] = W_n[0] * M[n]

# Compute beamformer output
Y_close = np.sum(M_filtered_close)
P_close = np.abs(Y_close)**2

print(f"Beamformer output: Y = {Y_close}")
print(f"  Magnitude: |Y| = {np.abs(Y_close):.2f}")
print(f"  Phase: ∠Y = {np.degrees(np.angle(Y_close)):.1f}°")
print(f"POWER: P({theta0_close}°) = {P_close:.2f}")

# ========== Test (ii): Steering angle FAR from source ==========
print(f"\n" + "="*70)
print(f"(ii) STEERING FAR FROM SOURCE: θ₀ = {theta0_far}°")
print("="*70)

# Compute filters and apply them for theta0_far
M_filtered_far = np.zeros(N, dtype=complex)
for n in range(N):
    W_n = beam_filter(antenne, np.array([f0_detected]), theta0=theta0_far, mic_nb=n)
    M_filtered_far[n] = W_n[0] * M[n]

# Compute beamformer output
Y_far = np.sum(M_filtered_far)
P_far = np.abs(Y_far)**2

print(f"Beamformer output: Y = {Y_far}")
print(f"  Magnitude: |Y| = {np.abs(Y_far):.2f}")
print(f"  Phase: ∠Y = {np.degrees(np.angle(Y_far)):.1f}°")
print(f"POWER: P({theta0_far}°) = {P_far:.2f}")

# ========== Comparison ==========
print(f"\n" + "="*70)
print("COMPARISON:")
print("="*70)
print(f"P({theta0_close}°) = {P_close:.2f}  (close to source)")
print(f"P({theta0_far}°) = {P_far:.2f}   (far from source)")
print(f"\nRatio: P(close)/P(far) = {P_close/P_far:.2f}")
print(f"Difference: P(close) - P(far) = {P_close - P_far:.2f}")
print("="*70)

if P_close > P_far:
    print(f"✓ P({theta0_close}°) > P({theta0_far}°) as expected!")
    print(f"  The beamformer correctly detects higher power when steered")
    print(f"  toward the actual source direction.")
else:
    print(f"✗ Unexpected result: P({theta0_far}°) > P({theta0_close}°)")
    print(f"  This might indicate the source is actually closer to {theta0_far}°")

# ========== Visualization ==========
fig, axes = plt.subplots(2, 2, figsize=(14, 10))

colors = ['red', 'blue', 'green', 'orange', 'purple', 'brown', 'pink', 'gray']

# Top row: Complex plane representation
# Plot (i): Close to source
for i in range(N):
    axes[0, 0].plot(np.real(M_filtered_close[i]), np.imag(M_filtered_close[i]), 
                   'o', markersize=10, color=colors[i], label=f'Mic {i}', alpha=0.7)
    axes[0, 0].arrow(0, 0, np.real(M_filtered_close[i]), np.imag(M_filtered_close[i]), 
                    head_width=np.abs(M_filtered_close[i])*0.03, 
                    head_length=np.abs(M_filtered_close[i])*0.05,
                    fc=colors[i], ec=colors[i], alpha=0.5, linewidth=1.5)

axes[0, 0].plot(np.real(Y_close), np.imag(Y_close), 's', 
               markersize=15, color='black', label='Sum', zorder=10)
axes[0, 0].arrow(0, 0, np.real(Y_close), np.imag(Y_close), 
                head_width=np.abs(Y_close)*0.03, head_length=np.abs(Y_close)*0.05,
                fc='black', ec='black', alpha=0.8, linewidth=3, zorder=9)

axes[0, 0].set_xlabel('Real Part', fontsize=11)
axes[0, 0].set_ylabel('Imaginary Part', fontsize=11)
axes[0, 0].set_title(f'θ₀ = {theta0_close}° (CLOSE to source)\n|Y| = {np.abs(Y_close):.1f}', 
                    fontsize=12, fontweight='bold')
axes[0, 0].grid(True, alpha=0.3)
axes[0, 0].axhline(y=0, color='k', linewidth=0.5)
axes[0, 0].axvline(x=0, color='k', linewidth=0.5)
axes[0, 0].legend(loc='best', fontsize=8)
axes[0, 0].set_aspect('equal')

# Plot (ii): Far from source
for i in range(N):
    axes[0, 1].plot(np.real(M_filtered_far[i]), np.imag(M_filtered_far[i]), 
                   'o', markersize=10, color=colors[i], label=f'Mic {i}', alpha=0.7)
    axes[0, 1].arrow(0, 0, np.real(M_filtered_far[i]), np.imag(M_filtered_far[i]), 
                    head_width=np.abs(M_filtered_far[i])*0.03, 
                    head_length=np.abs(M_filtered_far[i])*0.05,
                    fc=colors[i], ec=colors[i], alpha=0.5, linewidth=1.5)

axes[0, 1].plot(np.real(Y_far), np.imag(Y_far), 's', 
               markersize=15, color='black', label='Sum', zorder=10)
axes[0, 1].arrow(0, 0, np.real(Y_far), np.imag(Y_far), 
                head_width=np.abs(Y_far)*0.03, head_length=np.abs(Y_far)*0.05,
                fc='black', ec='black', alpha=0.8, linewidth=3, zorder=9)

axes[0, 1].set_xlabel('Real Part', fontsize=11)
axes[0, 1].set_ylabel('Imaginary Part', fontsize=11)
axes[0, 1].set_title(f'θ₀ = {theta0_far}° (FAR from source)\n|Y| = {np.abs(Y_far):.1f}', 
                    fontsize=12, fontweight='bold')
axes[0, 1].grid(True, alpha=0.3)
axes[0, 1].axhline(y=0, color='k', linewidth=0.5)
axes[0, 1].axvline(x=0, color='k', linewidth=0.5)
axes[0, 1].legend(loc='best', fontsize=8)
axes[0, 1].set_aspect('equal')

# Bottom row: Phase alignment visualization
# Plot phases before and after filtering
axes[1, 0].plot(range(N), np.degrees(np.angle(M_filtered_close)), 'o-', 
               markersize=10, linewidth=2, color='green', label='Phases after filtering')
axes[1, 0].set_xlabel('Microphone Number', fontsize=11)
axes[1, 0].set_ylabel('Phase (degrees)', fontsize=11)
axes[1, 0].set_title(f'Phase Alignment at θ₀ = {theta0_close}° (CLOSE)', 
                    fontsize=12, fontweight='bold')
axes[1, 0].grid(True, alpha=0.3)
axes[1, 0].set_xticks(range(N))
axes[1, 0].legend(fontsize=10)
axes[1, 0].axhline(y=0, color='black', linestyle='--', linewidth=1, alpha=0.5)

axes[1, 1].plot(range(N), np.degrees(np.angle(M_filtered_far)), 'o-', 
               markersize=10, linewidth=2, color='red', label='Phases after filtering')
axes[1, 1].set_xlabel('Microphone Number', fontsize=11)
axes[1, 1].set_ylabel('Phase (degrees)', fontsize=11)
axes[1, 1].set_title(f'Phase Alignment at θ₀ = {theta0_far}° (FAR)', 
                    fontsize=12, fontweight='bold')
axes[1, 1].grid(True, alpha=0.3)
axes[1, 1].set_xticks(range(N))
axes[1, 1].legend(fontsize=10)
axes[1, 1].axhline(y=0, color='black', linestyle='--', linewidth=1, alpha=0.5)

plt.tight_layout()
plt.show()

# Power comparison bar plot
fig, ax = plt.subplots(figsize=(10, 6))

powers = [P_close, P_far]
angles = [f'{theta0_close}°\n(CLOSE)', f'{theta0_far}°\n(FAR)']
bar_colors = ['green', 'red']

bars = ax.bar(range(2), powers, color=bar_colors, alpha=0.7, edgecolor='black', width=0.6)
ax.set_ylabel('Power P(θ₀)', fontsize=13)
ax.set_xlabel('Steering Angle', fontsize=13)
ax.set_title('Beamformer Power Comparison', fontsize=14, fontweight='bold')
ax.set_xticks(range(2))
ax.set_xticklabels(angles, fontsize=11)
ax.grid(True, alpha=0.3, axis='y')

# Add value labels
for i, (p, angle) in enumerate(zip(powers, angles)):
    ax.text(i, p, f'{p:.1f}', ha='center', va='bottom', fontsize=12, fontweight='bold')

plt.tight_layout()
plt.show()

print("\n" + "="*70)
print("CONCLUSION:")
print("="*70)
print("When the beamformer is steered CLOSE to the source direction:")
print("  • Phases are ALIGNED → vectors point in similar directions")
print("  • CONSTRUCTIVE interference → large sum magnitude")
print("  • HIGH POWER")
print("")
print("When the beamformer is steered FAR from the source direction:")
print("  • Phases are SCATTERED → vectors point in different directions")
print("  • DESTRUCTIVE interference → small sum magnitude")
print("  • LOW POWER")
print("")
print("This is how beamforming localizes sound sources!")
print("="*70)
```

### 11) Step (g) : Repeat now the previous code in a loop for $\theta_0$ values ranging from 0 to 180° .You should then obtain an array $\texttt{P}$ where each value corresponds to the power of the beamformer output at $F_0$ for each angular polarization. Plot the array $\texttt{P}$ as a function of the angle $\theta_0$.

```python
import numpy as np
import matplotlib.pyplot as plt

# Define range of steering angles
theta_array = np.arange(0, 181, 1)  # 0° to 180° with 1° step
P = np.zeros(len(theta_array))

print(f"Computing beamformer power for {len(theta_array)} angles...")

# Loop over all steering angles
for idx, theta0 in enumerate(theta_array):
    # Apply filters and compute beamformer output
    M_filtered = np.zeros(N, dtype=complex)
    for n in range(N):
        W_n = beam_filter(antenne, np.array([f0_detected]), theta0=theta0, mic_nb=n)
        M_filtered[n] = W_n[0] * M[n]
    
    # Compute power
    Y_theta0 = np.sum(M_filtered)
    P[idx] = np.abs(Y_theta0)**2

# Find peak
peak_idx = np.argmax(P)
estimated_angle = theta_array[peak_idx]
max_power = P[peak_idx]

print(f"Done! Peak at θ = {estimated_angle}° with power P = {max_power:.2f}")

# Plot
plt.figure(figsize=(12, 6))
plt.plot(theta_array, P, linewidth=2, color='blue')
plt.axvline(x=estimated_angle, color='red', linestyle='--', linewidth=2, 
            label=f'Peak at θ = {estimated_angle}°')
plt.xlabel('Steering Angle θ₀ (degrees)', fontsize=13)
plt.ylabel('Power P(θ₀)', fontsize=13)
plt.title(f'Beamformer Power Spectrum (f₀ = {f0_detected:.0f} Hz)', 
          fontsize=14, fontweight='bold')
plt.grid(True, alpha=0.3)
plt.xlim([0, 180])
plt.legend(fontsize=11)
plt.tight_layout()
plt.show()

print(f"\nEstimated source direction: {estimated_angle}°")
```
