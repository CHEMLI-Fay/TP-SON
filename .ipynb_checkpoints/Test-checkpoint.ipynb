```python
# Questions 19-21: FINAL CORRECTED VERSION
# Coordinate system: Z-axis horizontal (microphone array), X-axis vertical (up)
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import linregress
from scipy.optimize import differential_evolution

# Setup
d = 0.06  # 6cm spacing
c = 343   # m/s
frequency = 350  # Hz
N_mics = 8

# REAL SOURCE POSITION
# Z is HORIZONTAL (where mics are), X is VERTICAL (up)
z_s_real = 0.37  # 15cm horizontal
x_s_real = 0.215 # 48cm vertical (up)
r_s_real = np.sqrt(x_s_real**2 + z_s_real**2)

# θ_s is angle from Z-axis (horizontal) towards X-axis (up)
theta_s_real = np.degrees(np.arctan(x_s_real / z_s_real))

print("="*60)
print("COORDINATE SYSTEM:")
print("  Z-axis: Horizontal (microphone array, left to right)")
print("  X-axis: Vertical (upward)")
print("="*60)
print(f"Real source position:")
print(f"  z = {z_s_real*100:.0f}cm (horizontal)")
print(f"  x = {x_s_real*100:.0f}cm (vertical/up)")
print(f"  r = {r_s_real:.3f}m ({r_s_real*100:.1f}cm)")
print(f"  θ_s = {theta_s_real:.2f}° (from Z-axis towards X-axis)")
print()

# Read signal
m = antenne.read()

# Microphone positions on Z-axis (horizontal)
z_n = np.array([-0.21, -0.15, -0.09, -0.03, 0.03, 0.09, 0.15, 0.21])  # meters

print("Microphone positions on Z-axis (horizontal):")
for i in range(N_mics):
    print(f"  Mic {i+1}: z = {z_n[i]*100:+6.1f} cm")
print()

# ==================== Q19 ====================
print("Question 19: Time delays δτ_n relative to center O")
print("="*60)

# Use center (between mic 4 and 5) as reference
center_signal = (m[:, 3] + m[:, 4]) / 2

delays = np.zeros(N_mics)
for i in range(N_mics):
    corr = np.correlate(m[:, i], center_signal, mode='full')
    lag = np.argmax(corr) - (len(center_signal) - 1)
    delays[i] = lag / Fs

delays_ms = delays * 1000

for i in range(N_mics):
    print(f"Mic {i+1} at z={z_n[i]*100:+6.1f}cm: δτ_{i+1} = {delays_ms[i]:+8.4f} ms")

# Calculate d_n from delays
d_n = c * delays
d_n_cm = d_n * 100

print("\nExtra distances d_n = c × δτ_n:")
for i in range(N_mics):
    print(f"  Mic {i+1}: d_{i+1} = {d_n_cm[i]:+7.3f} cm")

# ==================== Q20 ====================
print("\nQuestion 20: Wavefront analysis")
print("="*60)

fig = plt.figure(figsize=(15, 6))
ax1 = plt.subplot(1, 2, 1)

# Plot delays vs position
for i in range(N_mics):
    ax1.plot(z_n[i]*100, delays_ms[i], 'o', markersize=10, color='steelblue')
    ax1.text(z_n[i]*100, delays_ms[i]+0.03, f'M{i+1}', 
             ha='center', va='bottom', fontsize=9, fontweight='bold')

ax1.plot(z_n*100, delays_ms, '-', linewidth=2, color='steelblue', 
         alpha=0.6, label='Measured delays')

ax1.set_xlabel('Microphone position z (cm) - Horizontal', fontsize=12)
ax1.set_ylabel('Delay δτ_n (ms)', fontsize=12)
ax1.set_title('Delay vs Position', fontsize=13, fontweight='bold')
ax1.grid(True, alpha=0.3)
ax1.axhline(0, color='k', linestyle='--', alpha=0.3, linewidth=1)
ax1.axvline(0, color='r', linestyle='--', alpha=0.4, linewidth=1.5, label='Center O')

# Linear fit
slope, intercept, r_value, _, _ = linregress(z_n, delays)
z_fit = np.linspace(z_n[0], z_n[-1], 100)
delays_fit = slope * z_fit + intercept
ax1.plot(z_fit*100, delays_fit*1000, 'r--', linewidth=2.5, 
         label=f'Linear fit (R²={r_value**2:.4f})')
ax1.legend(fontsize=11)

print(f"Linear fit R² = {r_value**2:.4f}")
wavefront_type = "plane" if r_value**2 > 0.95 else "spherical"
print(f"→ Wavefront type: {wavefront_type.upper()}")

# ==================== Q21 ====================
print("\nQuestion 21: Source localization using preparation equations")
print("="*60)

ax2 = plt.subplot(1, 2, 2)

if wavefront_type == "plane":
    # PLANE WAVE
    sin_theta = -slope * c
    sin_theta = np.clip(sin_theta, -1, 1)
    theta_s_est = np.arcsin(sin_theta)
    theta_s_deg = np.degrees(theta_s_est)
    
    print(f"From plane wave equation: sin(θ_s) = {sin_theta:.4f}")
    print(f"Estimated angle: θ_s = {theta_s_deg:.2f}° (from Z-axis)")
    print(f"Real angle: θ_s = {theta_s_real:.2f}° (from Z-axis)")
    print(f"Error: {abs(theta_s_deg - theta_s_real):.2f}°")
    
    # Visualization (X vertical, Z horizontal)
    for i in range(N_mics):
        ax2.plot(z_n[i], 0, 'ko', markersize=12, zorder=5)  # Mics on Z-axis (horizontal)
        ax2.text(z_n[i], -0.08, f'M{i+1}', ha='center', va='top', 
                 fontsize=9, fontweight='bold')
    
    # Draw Z-axis (horizontal)
    ax2.plot([-0.3, 0.3], [0, 0], 'k-', linewidth=2, alpha=0.5, label='Z-axis (horizontal)')
    ax2.axvline(0, color='r', linestyle='--', alpha=0.3, linewidth=1)
    
    # Wave fronts
    theta_rad = np.radians(theta_s_deg)
    for offset in np.linspace(-0.6, 0.6, 10):
        z_w = np.linspace(-0.35, 0.35, 50)
        x_w = z_w * np.tan(theta_rad) + offset / np.cos(theta_rad)
        ax2.plot(z_w, x_w, 'b-', alpha=0.15, linewidth=1)
    
    # Direction arrow (from Z-axis upward to X-axis)
    arrow_len = 0.3
    arrow_z = arrow_len * np.cos(theta_rad)
    arrow_x = arrow_len * np.sin(theta_rad)
    ax2.arrow(0, 0, arrow_z, arrow_x, head_width=0.03, head_length=0.03,
              fc='red', ec='red', linewidth=3, 
              label=f'Estimated: θ={theta_s_deg:.1f}°', zorder=4)
    
    # Real source
    ax2.plot(z_s_real, x_s_real, 'g*', markersize=25, 
             label=f'Real source\nθ={theta_s_real:.1f}°', zorder=6)
    ax2.plot([0, z_s_real], [0, x_s_real], 'g--', linewidth=2, alpha=0.7)
    
    ax2.set_title('Plane Wave Approximation', fontsize=13, fontweight='bold')
    ax2.set_xlim(-0.35, 0.35)
    ax2.set_ylim(-0.15, 0.6)

else:
    # SPHERICAL WAVE - Use equation (1)
    # d_n² + 2*d_n*r_s = z_n² - 2*r_s*z_n*cos(θ_s)
    
    def cost_function(params):
        r_s, theta = params
        total_error = 0
        for i, z in enumerate(z_n):
            a = 1
            b = 2 * r_s
            c_eq = -(z**2 - 2*r_s*z*np.cos(theta))
            discriminant = b**2 - 4*a*c_eq
            
            if discriminant < 0:
                return 1e10
            
            d_n_pred = (-b + np.sqrt(discriminant)) / (2*a)
            tau_pred = d_n_pred / c
            tau_meas = delays[i]
            total_error += (tau_pred - tau_meas)**2
        
        return total_error
    
    # Global optimization
    print("Running global optimization...")
    bounds = [(0.1, 1.5), (np.radians(10), np.radians(89))]
    result = differential_evolution(cost_function, bounds, seed=42, maxiter=1000)
    
    r_s_est, theta_s_est = result.x
    theta_s_deg = np.degrees(theta_s_est)
    z_s_est = r_s_est * np.cos(theta_s_est)
    x_s_est = r_s_est * np.sin(theta_s_est)
    
    print(f"\nFrom spherical wave equation (1):")
    print(f"Estimated: θ_s={theta_s_deg:.2f}° (from Z-axis), r_s={r_s_est:.3f}m")
    print(f"Estimated position: z={z_s_est:.3f}m, x={x_s_est:.3f}m")
    print(f"Real: θ_s={theta_s_real:.2f}° (from Z-axis), r_s={r_s_real:.3f}m")
    print(f"Real position: z={z_s_real:.3f}m, x={x_s_real:.3f}m")
    print(f"Angle error: {abs(theta_s_deg-theta_s_real):.2f}°")
    print(f"Distance error: {abs(r_s_est-r_s_real)*100:.1f}cm")
    print(f"Position error: {np.sqrt((z_s_est-z_s_real)**2+(x_s_est-x_s_real)**2)*100:.1f}cm")
    
    # Check residuals
    print("\nDelay comparison:")
    for i in range(N_mics):
        z = z_n[i]
        a = 1
        b = 2 * r_s_est
        c_eq = -(z**2 - 2*r_s_est*z*np.cos(theta_s_est))
        disc = b**2 - 4*a*c_eq
        d_n_pred = (-b + np.sqrt(max(0, disc))) / (2*a) if disc >= 0 else 0
        pred_delay = d_n_pred / c
        print(f"  M{i+1}: meas={delays_ms[i]:+7.4f}ms, pred={pred_delay*1000:+7.4f}ms, err={abs(delays_ms[i]-pred_delay*1000):.4f}ms")
    
    # Visualization (X vertical up, Z horizontal)
    for i in range(N_mics):
        ax2.plot(z_n[i], 0, 'ko', markersize=12, zorder=5)
        ax2.text(z_n[i], -0.08, f'M{i+1}', ha='center', va='top', 
                 fontsize=9, fontweight='bold')
    
    # Draw Z-axis (horizontal)
    ax2.plot([-0.3, 0.3], [0, 0], 'k-', linewidth=2, alpha=0.5)
    ax2.axvline(0, color='r', linestyle='--', alpha=0.3, linewidth=1)
    
    # Estimated source
    ax2.plot(z_s_est, x_s_est, 'r*', markersize=25,
             label=f'Estimated\nθ={theta_s_deg:.1f}°, r={r_s_est*100:.0f}cm', zorder=6)
    ax2.plot([0, z_s_est], [0, x_s_est], 'r--', linewidth=2, alpha=0.7)
    
    # Real source
    ax2.plot(z_s_real, x_s_real, 'g*', markersize=25,
             label=f'Real\nθ={theta_s_real:.1f}°, r={r_s_real*100:.0f}cm', zorder=6)
    ax2.plot([0, z_s_real], [0, x_s_real], 'g--', linewidth=2, alpha=0.7)
    
    # Spherical wave fronts
    for rad in np.linspace(r_s_est*0.5, r_s_est*1.5, 8):
        theta_c = np.linspace(0, 2*np.pi, 100)
        ax2.plot(z_s_est + rad*np.cos(theta_c), x_s_est + rad*np.sin(theta_c),
                 'b-', alpha=0.15, linewidth=1)
    
    ax2.set_title('Spherical Wave (Eq. 1)', fontsize=13, fontweight='bold')
    ax2.set_xlim(-0.35, 0.35)
    ax2.set_ylim(-0.15, 0.7)

# Format
ax2.set_xlabel('z (m) - Horizontal', fontsize=12)
ax2.set_ylabel('x (m) - Vertical', fontsize=12)
ax2.text(-0.32, 0.02, '← M1', fontsize=9, color='gray')
ax2.text(0.24, 0.02, 'M8 →', fontsize=9, color='gray')
ax2.grid(True, alpha=0.3)
ax2.legend(fontsize=10, loc='upper right')
ax2.axis('equal')

plt.tight_layout()
plt.show()

# ==================== DIAGNOSTICS ====================
print("\n" + "="*60)
print("DIAGNOSTIC: Expected delays at real position")
print("="*60)

expected_delays_ms = []
for i in range(N_mics):
    # Distance from source to mic i
    dist_to_mic_i = np.sqrt((z_s_real - z_n[i])**2 + x_s_real**2)
    # Distance from source to origin
    dist_to_origin = r_s_real
    # Extra distance
    d_n_expected = dist_to_mic_i - dist_to_origin
    # Expected delay
    tau_expected = d_n_expected / c
    expected_delays_ms.append(tau_expected * 1000)
    print(f"  M{i+1}: expected={tau_expected*1000:+7.4f}ms, measured={delays_ms[i]:+7.4f}ms, diff={abs(tau_expected*1000-delays_ms[i]):.4f}ms")

rms_error = np.sqrt(np.mean((np.array(expected_delays_ms) - delays_ms)**2))
print(f"\nRMS error: {rms_error:.4f}ms (~{rms_error*c/10:.2f}cm)")

# ==================== SUMMARY ====================
print("\n" + "="*60)
print("SUMMARY")
print("="*60)
print(f"Coordinate system: Z horizontal (mics), X vertical (up)")
print(f"Array: 8 mics, spacing={d*100}cm, length={7*d*100}cm")
print(f"Frequency: {frequency}Hz")
print(f"Real source: z={z_s_real}m, x={x_s_real}m")
print(f"Real: θ_s={theta_s_real:.2f}°, r={r_s_real:.3f}m")
print(f"Wavefront: {wavefront_type.upper()}")
if wavefront_type == "spherical":
    print(f"Estimated: θ_s={theta_s_deg:.2f}°, r={r_s_est:.3f}m")
    print(f"Errors: Δθ={abs(theta_s_deg-theta_s_real):.2f}°, Δr={abs(r_s_est-r_s_real)*100:.1f}cm")
print("="*60)
```
